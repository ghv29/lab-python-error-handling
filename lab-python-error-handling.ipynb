````python
#1 
products = ["t-shirt", "mug", "hat", "book", "keychain"]
#2
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity.")
    return inventory
#3
def get_customer_orders(inventory):
    customer_orders = set()
    
    while True: 
        try:
            num_orders = int(input("How many products would you like to order'? "))
            if num_orders < 0:
                raise ValueError("Number of orders cannot be negative.")
            break
        except ValueError as error:
            print(f"Invalid input:{error}")

    for _ in range(num_orders):
        while True:
            product = input("Enter the product you want to order: ").strip().lower()
            if product not in inventory:
                print("Product not found in inventory. Please enter a valid product")
            elif inventory[product] <= 0:
                print("Sorry, this product is out of stock. Please choose another product.")
            else:
                customer_orders.add(product)
                break
   
    
    print("Products in customer orders:", customer_orders)
    return customer_orders


#4
def update_inventory(inventory,customer_orders):
    inventory = {item: inventory[item] - 1 if item in customer_orders else inventory[item] for item in inventory}

    inventory = {item: qty for item, qty in inventory.items() if qty > 0}
    for item, qty in inventory.items():
        print(f"{item}, Quantity Left: {qty}")

    return inventory

def price_calculator(customer_orders):
    total_price = 0.0

    for product in customer_orders:
        valid_price = False
        while not valid_price:
            try:
                price = float(input(f"Enter the price of {product}: "))
                if price < 0:
                    raise ValueError("Price cannot be negative.")
                total_price += price
                valid_price = True
            except ValueError as error:
                print(f"Invalid input: {error}")

    print(f"Total Price: {total_price}")
    return total_price


#5
def calculate_order_statistics(customer_orders,products):
    total_products_ordered = len(customer_orders)
    percentage_ordered = (total_products_ordered / len(products)) * 100
    order_status = (total_products_ordered, percentage_ordered)
    return order_status
#6
def print_order_statistics(order_statistics):
    print("Order Statistics")
    print("Total Products Ordered:", order_statistics[0])
    print("Percentage of Products Ordered:", order_statistics[1], "%")
#7
def print_updated_inventory(inventory):
    print("Updated Inventory:")
    for item, qty in inventory.items():
        print(f"{item}: {qty}")


inventory = initialize_inventory(products)
customer_orders = get_customer_orders(inventory)
order_statistics = calculate_order_statistics(customer_orders, products)
print_order_statistics(order_statistics)
print_updated_inventory(inventory)
price_calculator(customer_orders)
````
